```python
def outlier(df, col, z):
	return df[abs(df[col] - np.mean(df[col]))/np.std(df[col]) > z].index

def not_outlier(df, col, z):
	return df[abs(df[col] - np.mean(df[col]))/np.std(df[col]) <= z].index

```

```python
game.loc[outlier(game, 'Global_Sales', 1.5)]
```

```python
game.loc[outlier(game, 'Global_Sales', 2)]
```

```python
game.loc[outlier(game, 'Global_Sales', 3)]
```

```python
game.loc[not_outlier(game, 'Global_Sales', 1.5)]
```
